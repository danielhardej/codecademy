```python
def convert_dict_to_table(filename):
    insurance_data_table = {}
    
    csv_file = open(filename)
        
    ages_list = []
    sex_list = []
    bmi_list = []
    children_list = []
    smoker_list = []
    region_list = []
    charges_list = []
    
    csv_dict = csv.DictReader(csv_file)

    for row in csv_dict:
        ages_list.append(row["age"])
        sex_list.append(row["sex"])
        bmi_list.append(row["bmi"])
        children_list.append(row["children"])
        smoker_list.append(row["smoker"])
        region_list.append(row["region"])
        charges_list.append(row["charges"])

    for i in range(len(ages_list)):
        new_record_dict = {}
        new_record_dict["age"] = ages_list[i]
        new_record_dict["sex"] = sex_list[i]
        new_record_dict["bmi"] = bmi_list[i]
        new_record_dict["children"] = children_list[i]
        new_record_dict["smoker"] = smoker_list[i]
        new_record_dict["region"] = region_list[i]
        new_record_dict["charges"] = charges_list[i]
        
        insurance_data_table[i] = new_record_dict
    print(insurance_data_table)
    return insurance_data_table


convert_dict_to_table("insurance.csv")

# for record in insurance_data:
#     print(record)
```
